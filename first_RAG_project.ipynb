```python
# This is our own small dataset of Random Facts - our "papers" for the filing cabinet
documents = [
    "The cheetah is the fastest land animal and can run up to 70 miles per hour.",
    "Penguins are birds that cannot fly but are excellent swimmers.",
    "Python is a popular programming language often used for beginners because it is easy to read.",
    "The solar system has eight planets, and Jupiter is the largest one.",
    "Water boils at 100 degrees Celsius at sea level.",
    "The Great Wall of China is the longest wall in the world and can be seen from space.",
    "Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne in 1976.",
    "The human body has 206 bones, and the smallest bone is in the ear.",
    "The currency used in Saudi Arabia is the Saudi Riyal (SAR).",
    "The currency used in Pakistan is the Pakistani Rupee (PKR).",
    "The worlds deadliest animal is a mosquito because 725,000 people are killed each year from mosquito-borne diseases.",
    "The loudest animal in the world is Pistol Shrimp which is capable of producing a sonic blast, louder than a Concorde’s sonic boom.",
    "Hummingbirds are the only known birds that can also fly backwards.",
    "Koalas can sleep for up to 22 hours in a day.",
    "Alexandra Fleming discovered Penicillin.",
    "Mars is known as the Red Planet.",
    "The most common name in the world is Muhammed.",
    "Bananas contain a naturally radioactive form of potassium, though not enough to be harmful.",
    "A pomegranate can hold up to 1400 seeds.",
    "Saturn is the only planet in our solar system that is less dense than water, meaning it could float in a large enough bathtub.",
    "The liver is the only human organ that can fully regenerate itself.",
    "A fear of long words is called Hippopotomonstrosesquippedaliophobia.",

]
```

We will use:

1. LLM (The Robot Brain): Groq API. It's super fast and is perfect for beginners.

2. Vector Database (The Filing Cabinet): ChromaDB. It's like a simple database that understands the meaning of words, not just the words themselves. It's very easy to set up.

3. Embeddings (The Magic Translator): A free model from Hugging Face (sentence-transformers/all-MiniLM-L6-v2). This model's job is to take our sentences and turn them into a list of numbers (a vector). The database uses these numbers to find similar sentences.

```python
# Importing our tools
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer

# 1. Start the magic translator
print("Loading the magic translator (embedding model)...")
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Create a new, empty filing cabinet (ChromaDB)
print("Creating the filing cabinet...")
chroma_client = chromadb.Client(Settings(allow_reset=True))
chroma_client.reset()

# 3. Create a drawer in the cabinet called "my_rag_collection"
collection = chroma_client.create_collection(name="my_rag_collection")

# 4. Translate our documents into numbers (embeddings) and put them in the cabinet
print("Putting papers into the filing cabinet...")
embeddings = embedder.encode(documents).tolist()
ids = [f"doc_{i}" for i in range(len(documents))] # Create IDs: doc_0, doc_1, etc.

# Now, add everything to the cabinet!
collection.add(
    documents=documents,
    embeddings=embeddings,
    ids=ids
)

print("Done! Your filing cabinet is ready with", len(documents), "papers.")
```

```python
# 1. Let's ask a question!
query = "Are bananas actually radioactive?"

# 2. Translate the question
print("Translating your question...")
query_embedding = embedder.encode([query]).tolist()

# 3. Search the filing cabinet
print("Searching the filing cabinet for the best answer...")
results = collection.query(
    query_embeddings=query_embedding,
    n_results=1
)

# 4. See what it found!
retrieved_doc = results['documents'][0][0]
print("\n🤖 The most relevant paper I found says:")
print('"' + retrieved_doc + '"')
```

```python
# THE COMPLETE RAG MACHINE!
from groq import Groq

client = Groq()
user_question = input("What would you like to know? \n> ")

print("\n🔍 Searching my knowledge for the answer...")
query_embedding = embedder.encode([user_question]).tolist()
results = collection.query(query_embeddings=query_embedding, n_results=1)
retrieved_doc = results['documents'][0][0]

print("\n📄 I found this information to help you:")
print(f'"{retrieved_doc}"')
print("\n🤖 Thinking of the best answer for you...\n")

system_prompt = """Use the following context to answer the question. If the answer is not in the context, say you don't know.

Context:
{context}
"""

chat_completion = client.chat.completions.create(
    messages=[
        {"role": "system", "content": system_prompt.format(context=retrieved_doc)},
        {"role": "user", "content": user_question}
    ],
    model="llama3-8b-8192", # Changed model to a currently supported one
    temperature=0.3
)

final_answer = chat_completion.choices[0].message.content
print("✅ Answer:")
print(final_answer)
```
